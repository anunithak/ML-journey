```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("cafe_cleaned.csv")
```

```python
##we're gonna be looking at variance, standard deviation and visualising it with a histogram
```

```python
col=df["Total Spent"]
variance=col.var()
```

```python
variance
```

```python
standard_deviation=col.std()
standard_deviation
```

```python
plt.hist(col, bins=50, edgecolor='black')
plt.title(f"Distribution of Total Spent")
plt.xlabel("Total Spent")
plt.ylabel("Frequency")
plt.show()
```

```python

##now lets compare standard deviation of total spent per item
std_by_item = df.groupby("Item")["Total Spent"].std()
print(std_by_item)

##it tells you which items have more variability in spending
```

```python
##lets check if the data is skewed 

from scipy.stats import skew

skewness = skew(col.dropna())  # drop NaNs
print("Skewness:", skewness)

if skewness > 0:
    print("Right-skewed (long tail on the high end)")
elif skewness < 0:
    print("Left-skewed (long tail on the low end)")
else:
    print("Symmetric distribution")
```
