```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("titanic.csv")
```

```python
plt.hist(df['age'].dropna(), bins=20, color='skyblue', edgecolor='black') 
##bins=20 would divide all the ages into 20 equal groups 
##dropna removes rows or columns with missing values
plt.title("Age Distribution of Titanic Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
survival_by_gender = df.groupby('sex')['survived'].mean()

plt.bar(survival_by_gender.index, survival_by_gender.values, color=['lightblue', 'pink'])
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.show()
```

```python
survival_by_class = df.groupby('pclass')['survived'].mean()

plt.bar(survival_by_class.index, survival_by_class.values, color='orange')
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

```python
plt.scatter(df['age'], df['fare'], alpha=0.5)
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
