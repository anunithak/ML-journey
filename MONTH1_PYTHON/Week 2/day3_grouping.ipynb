```python
import pandas as pd
```

```python
df=pd.read_csv("titanic.csv")
```

```python
##Group Titanic by Pclass
##Compute average Age and survival rate per class

df.groupby("pclass")[["age","survived"]].mean()
```

```python
##Group by Sex
##Count passengers per gender.

df.groupby("sex").size()
```

```python
##Pivot Table:
##Show average Fare by Pclass and Sex.

df.pivot_table(values="fare",index="pclass",columns="sex",aggfunc="mean")
```
