```python
import pandas as pd
from scipy import stats
import numpy as np
```

```python
df=pd.read_csv("covid_cleaned.csv")
```

```python
df.duplicated().sum()
```

```python
##removing the duplicates

df.drop_duplicates(inplace=True)
```

```python
##well i guess there's no duplicate values at all
##still doing the stuff thats supposed to be done for today
```

```python
df.info()
```

```python
# CORRECTED syntax:
cols = ["Total Cases","Total Deaths","Total Recovered","Active Cases","Total Test","Population"]
df[cols] = (df[cols]
    .replace(',', '', regex=True)
    .replace(['unknown', 'N/A', '-'], np.nan)
    .astype(float))
```

```python
df.info()
```

```python
z_scores=stats.zscore(df["Total Cases"].dropna())
outliers=df[np.abs(z_scores)>3]
```

```python
print(outliers[["Country","Total Cases"]])
```

```python
valid = df["Total Deaths"].dropna()
z_scores2=stats.zscore(valid)
mask = np.abs(z_scores2) > 3
outliers2 = df.loc[valid.index[mask], ["Country", "Total Deaths"]]
print(outliers2)
```
