```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("cafe_cleaned.csv")
```

```python
corr_matrix=df[["Total Spent","Quantity","Price Per Unit"]].corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix,annot=True,cmap="coolwarm")
plt.show()
```

```python
## lets identify two positive correlations(highest positive numbers that arent 1.0 coz thats self correlation)
##and two negative correlations (lowest negative numbers)
##there are no negative correlations 

##here the top positive relations are :
##1. quantity vs Total spent (0.7)- which makes sense as buying more means it costs more money , when people buy more units their total spending increases
##2. total spent vs price per unit (0.6)- again very obvious, if the price per unit is high, the total amount spent increases


##for negative relations, since there's none: there's one almost close to zero
### One surprising finding is that we've found almost no correlation between quantity and price per unit
##suggesting that no matter the price, people are buying similar quantities
##so price hasnt affected purchase quantity

```
