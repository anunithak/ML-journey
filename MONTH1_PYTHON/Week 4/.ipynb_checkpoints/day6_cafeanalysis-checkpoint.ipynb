```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("cafe_cleaned.csv")
```

```python
df.describe()
```

```python
corr_matrix=df[["Total Spent","Quantity","Price Per Unit"]].corr()
```

```python
sns.heatmap(corr_matrix,annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
price_stats = df.groupby("Item")["Price Per Unit"].agg(["mean","median","count"])
print(price_stats)
```

```python
df["Item"].value_counts()
```

```python
avg_total_spent = df.groupby("Item")["Total Spent"].mean()
print("\nAverage spending per item:\n", avg_total_spent)
```

```python
total_revenue = df.groupby("Item")["Total Spent"].sum()
print("\nTotal revenue per item:\n", total_revenue)
```

```python
total_revenue.plot(kind="bar")
plt.title("Total Revenue by Item")
plt.ylabel("Revenue")
plt.show()
```

```python
df["Payment Method"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Payment Method Split")
plt.ylabel("")
plt.show()
```

```python
df["Total Spent"].plot(kind="hist", bins=10)
plt.title("Distribution of Total Spent")
plt.xlabel("Total Spent")
plt.show()
```

## Summary


#### This analysis examines customer purchasing behaviour in a cafe using descriptive statistics, correlations, groupby operations and multiple visualisations

## Correlation Analysis

#### Total Spent and Quantity show a strong positive correlation, meaning most of the spending variation is driven by how many items a customer buys.

#### Price Per Unit and Quantity have almost zero correlation, suggesting customers purchase similar quantities regardless of price.

## Revenue Insights

#### Salad produces the highest total revenue, driven by its higher price point.

#### Sandwich and Smoothie are the next highest contributors.

#### Coffee and Cookies, although frequently purchased, generate lower revenue due to lower unit prices.

## Payment Method Trends

#### The caf sees an almost even split across Cash, Credit Card, and Digital Wallet, indicating diverse customer preferences and balanced POS usage.

## Overall  Conclusions

#### Customers do not change purchase quantity based on price.

#### Higher-priced items drive the majority of revenue, even if purchased less often.
