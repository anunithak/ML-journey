```python
import pandas as pd
```

```python
df=pd.read_csv("covid.csv")
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df.isnull().sum()
```

```python
## trying to fill the null values with the mean of the populations
df["Population"] = pd.to_numeric(df["Population"], errors="coerce")
df["Population"] = df["Population"].fillna(df["Population"].mean())
```

```python
df.isnull().sum()
```

```python
df["Total Deaths"] = pd.to_numeric(df["Total Deaths"], errors="coerce")
df["Total Deaths"]=df["Total Deaths"].fillna(df["Total Deaths"].median())
```

```python
df["Total Recovered"] = pd.to_numeric(df["Total Recovered"], errors="coerce")
df["Total Recovered"]=df["Total Recovered"].fillna(df["Total Recovered"].median())
```

```python
df["Total Cases"] = df["Total Cases"].astype(str).str.replace(",", "").astype(float)
df["Total Test"] = df["Total Test"].astype(str).str.replace(",", "").astype(float)
```

```python
df.isnull().sum()
```

```python
df.to_csv("covid_cleaned.csv", index=False)
```
