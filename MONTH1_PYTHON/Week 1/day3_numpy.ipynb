```python
import numpy as np
print(np.__version__)
print(np.array([1, 2, 3]))
```

```python
##creating an array with numbers from 1-10 and then only printing the even numbers
```

```python
num=np.arange(1,11)
for i in range(10):
    if num[i]%2 ==0:
        print (num[i])
```

```python
##or a more professional way is by using boolean masking
```

```python
even_nums=num[num%2==0]
print(even_nums)
```

```python
##here you're basically asking a true or false question
##by using the boolean array
##so when you do this you're basically saying give me those elements of num where
##num %2==0  is true, and numpy masks out everything else thats false
```

```python
##NOW SIMULATING 1000 DICE ROLLS, PRINT AVERAGE ROLLS AND STANDARD DEVIATION
```

```python
rand100=np.random.randint(1,7,1000)
print(np.mean(rand100))
print(np.std(rand100))
```

```python
##now creating a 3x3 random matrix and computing the sum of each row and column
```

```python
matr3=np.random.rand(3,3)
print(np.sum(matr3,axis=0))
print(np.sum(matr3,axis=1))
```

```python
## or to get proper integer random numbers instead of float
matr3_int=np.random.randint(0,10,size=(3,3))
print(np.sum(matr3_int,axis=0))
print(np.sum(matr3_int,axis=1))
```
