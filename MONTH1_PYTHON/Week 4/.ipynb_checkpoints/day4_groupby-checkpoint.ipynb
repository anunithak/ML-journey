```python
import pandas as pd
```

```python
df=pd.read_csv("cafe_cleaned.csv")

##now we'll groupby certain columns and find the following things
##the most expensive item
##how much is being spent in store and take away
##which item sells the most
```

```python
df.groupby("Item")["Price Per Unit"].mean().sort_values(ascending=False)
```

```python
##here we see that salad is the most expensive item after sorting it
```

```python
df.groupby("Location")["Total Spent"].agg(["mean","median","count"])
```

```python
##we see that customers buy more in store than for takeaway
```

```python
df["Item"].value_counts()
```

```python
## juice is the most sold item , followed by coffee
```
