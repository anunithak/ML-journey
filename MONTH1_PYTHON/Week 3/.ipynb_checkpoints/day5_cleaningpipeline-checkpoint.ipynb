```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats 
```

```python
df=pd.read_csv("dirty_cafe_sales.csv")
df.isnull().sum()
```

```python
df.info()
```

```python
df.head(10)
```

```python
##quantity, price per unit and total spent need to be integers
##calculate the total spent using quantity and price per unit, also the same for quantity and price per unit
##set all unknown values to unknown
##check for duplicates and extreme outliers
```

```python
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
##since we have non numeric values like "ERROR" or "Unknown" we use errors="coerce" so that
##the non numeric values become NaN
```

```python
df["Price Per Unit"]=pd.to_numeric(df["Price Per Unit"],errors="coerce")
df["Total Spent"]=pd.to_numeric(df["Total Spent"],errors="coerce")
```

```python
df.info()
```

```python
df["Total Spent"].fillna(df["Quantity"]*df["Price Per Unit"],inplace=True)
```

```python
df.isnull().sum()
```

```python
df["Quantity"].fillna(df["Total Spent"]/df["Price Per Unit"],inplace=True)
df["Price Per Unit"].fillna(df["Total Spent"]/df["Quantity"],inplace=True)
```

```python
df.isnull().sum()
```

```python
##now lets try to replace all bad strings in the item and payment method columns
df['Item'].unique()
```

```python
##so the bad strings are "unknown" and "error"
```

```python
df["Payment Method"].unique()
df["Location"].unique()
```

```python
df["Item"].replace({"UNKNOWN":np.nan,"ERROR":np.nan},inplace=True)
df["Payment Method"].replace({"UNKNOWN":np.nan,"ERROR":np.nan},inplace=True)
df["Location"].replace({"UNKNOWN":np.nan,"ERROR":np.nan},inplace=True)
df["Transaction Date"].replace({"UNKNOWN":np.nan,"ERROR":np.nan},inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
z_scores = stats.zscore(df['Quantity'], nan_policy='omit') 

##we used this line because this -> z_scores = stats.zscore(df['Quantity'].dropna()) gave an error because 
##Whenever you remove NaNs from a column, the resulting Series is shorter.
##You cannot directly use that shorter array to filter the full dataframe,
##because Pandas expects the boolean array to be the same length as the dataframe.

outliers = df[np.abs(z_scores) > 3]  
```

```python
z_scores1 = stats.zscore(df['Price Per Unit'], nan_policy='omit') 
outliers1 = df[np.abs(z_scores) > 3]  
```

```python
z_scores2 = stats.zscore(df['Total Spent'], nan_policy='omit') 
outliers2 = df[np.abs(z_scores) > 3]  
```

```python
df['Payment Method'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.ylabel("")
plt.show()
```

```python
pd.crosstab(df['Item'], df['Location']).plot(kind='bar', stacked=True)
plt.ylabel("Count")
plt.show()
```

```python
df.groupby('Item')['Total Spent'].mean().plot(kind='bar')
plt.ylabel("Total amount spent (average)")
plt.show()
```
