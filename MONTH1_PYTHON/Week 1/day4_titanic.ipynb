```python
import pandas as pd
```

```python
df=pd.read_csv("titanic.csv")
```

```python
print(df.head(10)) 
```

```python
print(df.isnull().sum())
```

```python
print("Total passengers",len(df))
```

```python
print("Overall Survival Rate: ", df["survived"].mean())
```

```python
print("Survival rate by gender: ", df.groupby("sex")["survived"].mean())
```

```python
print("Average age by class: ",df.groupby("pclass")["age"].mean())
```

```python
df["AgeGroup"]=pd.cut(df["age"],bins=[0,12,18,60,100],labels=["Child","Teen","Adult","Senior"])
```

```python
print(df["AgeGroup"])
```

```python
print(df.groupby(["pclass","sex"])["survived"].mean())
```
