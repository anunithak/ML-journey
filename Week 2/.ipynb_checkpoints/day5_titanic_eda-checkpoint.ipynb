```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("titanic.csv")
```

```python
##DATA CLEANING workflow
##detecting missing values
df.isnull().sum()
```

```python
##Fill the missing values
##or drop missing rows
##replace inconsistent categorical data
##verify the changes 
```

```python
##Creating a new column Agegroup
df["AgeGroup"]=pd.cut(df["age"],bins=[0,12,18,60,100],labels=["Child","Teen","Adult","Senior"])
```

```python
df.info()
```

```python
##Analysing survival by gender, pclass, and agegroup
sns.barplot(x="sex",y="survived",data=df)
```

```python
survival_by_class=df.groupby("pclass")["survived"].mean()
plt.pie(survival_by_class.values,labels=survival_by_class.index,autopct='%1.1f%%')
```

```python
survival_by_agegroup=df.groupby("AgeGroup")["survived"].mean()
sns.heatmap(survival_by_agegroup.to_frame(),annot=True,fmt=".2f")
```
