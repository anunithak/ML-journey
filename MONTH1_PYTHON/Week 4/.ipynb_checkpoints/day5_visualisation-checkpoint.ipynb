```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("cafe_cleaned.csv")
```

```python
sns.barplot(x="Item",y="Price Per Unit",data=df)

##salad is the highest priced item, followed by smoothies and sandwiches
```

```python
df['Payment Method'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Payment Method Split")
plt.ylabel("")
plt.show()

##the payment methods are used almost equally 
```

```python
df.groupby('Item')['Total Spent'].sum().plot(kind='bar')
plt.title("Total Revenue by Item")
plt.ylabel("Total Spent ()")
plt.show()

##we get the most revenue from salad, and its also the highest priced item
```
